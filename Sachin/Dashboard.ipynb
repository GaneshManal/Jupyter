```python
def string_df_columns_to_numeric(df, column_list):
    df[column_list] = df[column_list].apply(pd.to_numeric, errors='test')
```

```python
import pandas as pd
df = pd.read_csv('odi_career.csv')
```

```python
string_df_columns_to_numeric(df, ["batting_score", "wickets"])
```

```python
%matplotlib inline
```

```python
batting_stats = df.groupby(['batting_innings'])['batting_score'].sum()
batting_stats.plot.bar(title="Score group by inning", style={})
```

```python
batting_stats = df.groupby(['opposition'])['batting_score'].sum()
batting_stats.plot.bar(color="purple", title="Run scored group by opposition", style={})
```

```python
batting_stats = df.groupby(['opposition'])['batting_score'].max()
batting_stats.plot.bar(color="green", title="Max Runs score against any opp", )
```

```python
batting_stats = df.groupby(['match_result'])['batting_score'].sum()
batting_stats.plot.barh(color="blue", title="Runs score group by match result", style={})
```
